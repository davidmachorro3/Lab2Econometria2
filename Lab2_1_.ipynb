4. Dados los arrays a = [1, 4, 2, 7] y b = [1, 3, 2, 9],
obtenga la media aritmética de la diferencia (a-b).

```python
import numpy as np

a = np.array([1, 4, 2, 7])
b = np.array([1, 3, 2, 9])

diferencia = a - b

media_aritmetica = np.mean(diferencia)

print("Media aritmetica: ", media_aritmetica)
```

2. Realice una gráfica de barras sobre la población de cada ciudad, ordenela de menor a mayor y responda:

a. ¿Cuál es la segunda ciudad más poblada?


b. ¿Qué posición ocupa Granada en el ranking de
las ciudades más pobladas?

```python
import matplotlib.pyplot as plt

ordenado = df.sort_values(by='Poblacion')

plt.figure(figsize=(10, 6))
plt.barh(ordenado['Municipio'], ordenado['Poblacion'], color='skyblue')
plt.xlabel('Poblacion')
plt.ylabel('Ciudad')
plt.title('Poblaciones')
plt.gca().invert_yaxis()
plt.show()

segundo_lugar = ordenado['Municipio'].iloc[-2]
print("Segunda mas poblada: ", segundo_lugar)

granada = ordenado[ordenado['Municipio'] == 'Granada'].index[0] + 1
print("Posicion Granada:", granada)
```

4.  ¿Cuál es el municipio situado más al Norte? (Usar el
valor de la coordenada "Y" que representa la latitud en
grados). Proporcione también la provincia a la que
pertenece y su población.

```python
mas_alto = df[df['Y'] == df['Y'].max()]

municipio = mas_alto['Municipio'].values[0]
provincia = mas_alto['Provincia'].values[0]
poblacion = mas_alto['Poblacion'].values[0]

print("Municipio mas al norte: ", municipio)
print("Provincia: ", provincia)
print("Poblacion: ", poblacion)
```

5. Obtenga la media, mediana, desviación estándar, valor
máximo y valor mínimo de la población de los municipios
de la provincia de Granada.

```python
munis = df[df['Provincia'] == 'Granada']

media = munis['Poblacion'].mean()
mediana = munis['Poblacion'].median()
desviacion = munis['Poblacion'].std()
max = munis['Poblacion'].max()
min = munis['Poblacion'].min()

print("Media: ", media)
print("Mediana: ", mediana)
print("Desviacion: ", desviacion)
print("Maximo: ", max)
print("Minimo: ", min)
```

6. Realice un histograma con la población de los
Municipios para cada una de las provincias.

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 8))
for provincia, datos in df.groupby('Provincia'):
    plt.hist(datos['Poblacion'], bins=20, alpha=0.5, label=provincia)

plt.xlabel('Poblacion')
plt.ylabel('Frecuencia')
plt.title('Poblacion/Provincia')
plt.legend()
plt.show()
```

8. Determine los nombres de los Municipios más cercano y
más lejano a Madrid. Para ello debe calcular la
distancia en todos ellos y Madrid. No considere a Madrid
en el análisis ya que la distancia sería cero.

```python
from geopy.distance import geodesic
madrid = (40.4168, -3.7038)

df['DistanciaConMadrid'] = df.apply(lambda row: geodesic(madrid, (row['Y'], row['X'])).kilometers, axis=1)

nearest = df[df['DistanciaConMadrid'] == df[df['Municipio'] != 'Madrid']['DistanciaConMadrid'].min()]['Municipio'].values[0]

farthest = df[df['DistanciaConMadrid'] == df['DistanciaConMadrid'].max()]['Municipio'].values[0]

print("Mas cercano: ", nearest)
print("Mas lejano: ", farthest)
```
